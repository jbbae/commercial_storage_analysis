```python
from sklearn.linear_model import SGDRegressor
X = df.as_matrix(columns=['hour', 'OAT'])
y = df['Building 6 kW'].tolist()

n_iter = np.ceil(10**6 / len(y))

clf = SGDRegressor(loss="squared_loss", penalty=None, max_iter=n_iter, tol=1e-3, random_state=42)
clf.fit(X, y)
```

```python
X_pred = pd.DataFrame({'hour': [13], 'OAT': [35]})

y_pred = clf.predict(X_pred)

print(y_pred)
```

```python
clf2 = SGDRegressor(penalty='l2', alpha=0.15, max_iter=n_iter, tol=1e-3, random_state=42)

clf2.fit(X,y)

X_pred2 = pd.DataFrame({'hour': [13], 'OAT': [35]})
y_pred2 = clf2.predict(X_pred2)

print(y_pred2)
```
