```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits import mplot3d
import datetime
%matplotlib inline

df = pd.read_csv('./data/building60preoffice.csv')
df.head()
df.shape
```

```python
from sklearn.linear_model import SGDRegressor
X = df.as_matrix(columns=['hour', 'OAT'])
y = df['Building 6 kW'].tolist()

n_iter = np.ceil(10**6 / len(y))

clf = SGDRegressor(loss="squared_loss", penalty=None, max_iter=n_iter, tol=1e-3, random_state=42)
clf.fit(X, y)
```

```python
X_pred = pd.DataFrame({'hour': [13], 'OAT': [35]})

y_pred = clf.predict(X_pred)

print(y_pred)
```

```python
clf2 = SGDRegressor(penalty='l2', alpha=0.15, max_iter=n_iter, tol=1e-3, random_state=42)

clf2.fit(X,y)

X_pred2 = pd.DataFrame({'hour': [13], 'OAT': [35]})
y_pred2 = clf2.predict(X_pred2)

print(y_pred2)
```

```python
df_w = pd.read_csv('./data/austin_weather.csv')
df_w.head()
```

```python
df_w2 = df_w.set_index("Date")

df_w2 = df_w2['2016-01-01':'2016-12-31']
df_w2['Month'] = df_w2.index.to_series().apply(lambda x: pd.to_datetime(x).month)

df_w2.head()
```

```python
sns.set(rc={'figure.figsize': (11,4)})

df_w2['TempHighF'].plot(linewidth=2);
df_w2['TempAvgF'].plot(linewidth=2);
df_w2['TempLowF'].plot(linewidth=2);
plt.title("Average Temperature (Austin, TX)")
# plt.savefig("total_listings_time.png")
```

```python
df_w2.index
```

```python
plt.plot(df_w2['Month'], df_w2['TempAvgF'], 'o', color='black');
```

```python
df_ausTempH = pd.DataFrame(columns=['Date', 'Hour', 'TempAvgF'])
```

```python
for index,row in df_w2.iterrows():
    for h in range(23):
        hrStr = str(h+1)
        if len(hrStr) == 1:
            hrStr = "0" + hrStr
        
        datetimeStr = index + " " + hrStr
        
        df_ausTempH = df_ausTempH.append({'Date': datetime.datetime.strptime(datetimeStr, '%Y-%m-%d %H'), 'Hour': h+1, 'TempAvgF': row['TempAvgF']}, ignore_index=True)
```

```python
print(df_ausTempH.shape)
df_ausTempH.head(48)
```

```python
# pd.to_datetime(df['Date'] + ' ' + df['Time'])

df_ausTempH['Date'] = pd.to_datetime(df_ausTempH['Date'])
# df_ausTempH['Date'] = df_ausTempH['Date'].replace(hour=df_ausTempH['Hour'])

# df_ausTempH['Date'] = df_ausTempH['Date'].dt.normalize() + datetime.timedelta(hours=df_ausTempH['Hour'])
df_ausTempH['Date'] = df_ausTempH['Date'].dt.normalize() + df_ausTempH['Hour'].map(datetime.timedelta)


df_ausTempH = df_ausTempH.set_index(pd.DatetimeIndex(df_ausTempH['Date']))
```
